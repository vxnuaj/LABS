```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [1, 2, 3] # controls the position of each bar,  based on it's alignment parameter
height = [3, 5, 3] # controls the height of each bar
width = .4

fig, axs = plt.subplots(1, 3, facecolor = 'orange')

fig.set_size_inches(w = (10, 7))

axs[0].bar(x, height, width,  align = 'edge', color = 'lightgreen') # each bar is on the right side of the `x` value
axs[1].bar(x, height, -width, align = 'edge', color = 'pink') # each bar is on the left side of the `x` value given the negative `width` param
axs[2].bar(x, height, width, align = 'center', color = 'lightblue') # each bar is at the center of the `x` value given the `align = 'center` (defualt setting)

plt.show()
```

```python
x = [1, 2, 3] # controls the position of each bar,  based on it's alignment parameter
height = [3, 5, 3] # controls the height of each bar
width = .4

fig, axs = plt.subplots(1, 3, facecolor = 'orange')

fig.set_size_inches(w = (10, 7))

axs[0].barh(x, height, width,  align = 'edge', color = 'lightgreen') # each bar is on the right side of the `x` value
axs[0].set_xlim(6, 0)
axs[1].barh(x, height, -width, align = 'edge', color = 'pink') # each bar is on the left side of the `x` value given the negative `width` param
axs[2].barh(x, height, width, align = 'center', color = 'lightblue') # each bar is at the center of the `x` value given the `align = 'center` (defualt setting)


plt.show()
```

```python
from time import time

start_time = time()

fig, axs = plt.subplots(2,  2)

x = [1, 2, 3, 4]
y = [1, 2, 3, 4]


for ax in axs.flat:
    ax.plot(x, y)
    ax.set_xticks([10, 20, 30, 40])
    ax.set_yticks([10, 20, 30, 40])


plt.show()

end_time = time()

print(f"Time Taken: {end_time - start_time} seconds")
```

```python
from time import time

start_time = time()

fig, axs = plt.subplots(2,  2)

x = [1, 2, 3, 4]
y = [1, 2, 3, 4]


for ax in axs.flatten():
    ax.plot(x, y)
    ax.set_xticks([10, 20, 30, 40])
    ax.set_yticks([10, 20, 30, 40])


plt.show()

end_time = time()

print(f"Time Taken: {end_time - start_time} seconds")
```

```python
data_dict = {'Cells': [1, 2, 3, 1], 'Cancerous':[1, 1, 0, 1]}
index =[f"Patient {i+1}" for i, _ in enumerate(data_dict['Cancerous'])]
df = pd.DataFrame(data_dict, index = index)

print(df)
```
