A Figure is where matplotlib paints our data. 

Within each Figure we can have multiple `Axes`, which represents the seperate areas that we actually plot our data onto. 

Each `Axes` has its seperate coordinate system, allowing us to describe how exactly we want to visually describe our data.

```python
fig, ax = plt.subplots(nrows = 2, ncols = 2) # Creating a figure with 4 axes (2x2 grid)

x = [1, 2, 3, 4]
y = [1, 4, 2, 3]

ax[0, 0].plot(x, y)

plt.show()
```

```python
del fig

fig, axs = plt.subplots(2, 2, layout = 'constrained') # automatically creates a figure (fig) and an ndarray with the corresponding # of Axes (ax).

# adding titles to the axes
axs[0, 0].set_title(f"Axes {1}", loc = 'left', fontstyle = 'italic') 
axs[0, 1].set_title(f"Axes {2}", loc = 'left', fontstyle = 'italic')
axs[1, 0].set_title(f"Axes {3}", loc = 'left', fontstyle = 'italic')
axs[1,1].set_title(f"Axes {4}", loc = 'left', fontstyle = 'italic')

# changing the color and title of the figure (and the font size of the tit el)
fig.set_facecolor('lightskyblue')
fig.suptitle('DING!', size = 'xx-large')

# getting total axes count
print(f"Total count of Axes in the figure: {axs.size}")

# plt.tight_layout() # making sure out axes / subplots don't overlap[] or just add layout = 'constrained' to plt.subplots
plt.show() # showing the figure.
```

You can easily specify mroe complex plots using `plt.subplot_mosaic`, where the input to the function is an array where each unique string denotes the position / size of subplots.

```python
del fig

#fig, axs = plt.subplot_mosaic([['Zeta', 'Zeta', 'Beta'],
#                               [ 'Alpha', 'Alpha', 'Beta']])

fig, axs = plt.subplot_mosaic('''
                              ZZB
                              ZZB
                              AAB
                              ''', layout = 'constrained', sharex = False, sharey = False)

# both produce the same effect

# plt.tight_layout() # if not using layout = 'constrained' in plt.subplot_mosaic

# DUMMY DATA
x = [1, 2, 3, 4]
y = [1, 4, 2, 3]

# Adding dummy data to each subplot
axs['Z'].plot(x, y)
axs['B'].plot(x, y)
axs['A'].plot(x, y)
axs['A'].sharey(axs['Z'])
axs['A'].sharex(axs['B'])
axs['Z'].sharex(axs['B'])

print(axs.items())
 
font_dict = {'fontsize': 'xx-large', 'fontweight': 'bold'} # **kwargs dict for Axes.text

for ax_name, ax in axs.items():
    ax.text(x = 2, y = 2, s = ax_name, **font_dict) # setting the text for each subplot / Axes

plt.show()
```

### You can create a figure and make subfigures within that figure

```python
del fig

fig = plt.figure(layout = 'constrained', facecolor = 'pink')
fig.suptitle('DING!', fontsize = 'xx-large', fontweight = 'bold')
left_fig, right_fig = fig.subfigures(1, 2)
right_fig.set_facecolor('lightblue')
left_fig.set_facecolor('turquoise')

x = [1, 2, 3, 4]
y = [1, 2, 3, 4]

left_ax1 = left_fig.add_subplot(2, 2, 1)
left_ax2 = left_fig.add_subplot(2, 2, 2)
left_ax3 = left_fig.add_subplot(2, 2, 3)
left_ax4 = left_fig.add_subplot(2, 2, 4)

right_ax1 = right_fig.add_subplot(2, 2, 1)
right_ax2 = right_fig.add_subplot(2, 2, 2)
right_ax3 = right_fig.add_subplot(2, 2, 3)
right_ax4 = right_fig.add_subplot(2, 2, 4)

left_axs = left_fig.get_axes()
right_axs = right_fig.get_axes()

for i, ax in enumerate(left_axs):
    ax.text(x = .5, y = .5, va = 'center', ha = 'center', s = f'Left {i + 1}')
    
for i, ax in enumerate(right_axs):
    ax.text(x = .5, y = .5, va = 'center', ha = 'center', s = f'Right {i + 1}')

print(fig.get_axes())

plt.show()
```

```python
del fig

fig = plt.figure(layout = 'constrained', facecolor='pink')
fig.suptitle(t = 'DING!', fontweight = 'bold', fontsize = 'xx-large')

left_fig, right_fig = fig.subfigures(nrows = 1, ncols = 2)

left_fig.add_subplot(2,2,1)
left_fig.add_subplot(2,2,2)
left_fig.add_subplot(2,2,3)
left_fig.add_subplot(2,2,4)
left_fig.set_facecolor('turquoise')

right_fig.add_subplot(2,2,1)
right_fig.add_subplot(2,2,2)
right_fig.add_subplot(2,2,3)
right_fig.add_subplot(2,2,4)
right_fig.set_facecolor('lightblue')

fontdict = {'ha': 'center',  'va': 'center', 'usetex':True}
x, y = .5, .5


for i, ax in enumerate(left_fig.get_axes()):
    ax.text(x, y, s = f'$\sum{i}$', **fontdict)

for i, ax in enumerate(right_fig.get_axes()):
    ax.text(x, y, s = f'$\sum{i}$', **fontdict)


plt.show()
```

```python
plt.ioff()

fig, axs = plt.subplot_mosaic('''
                   AAB
                   DDC
                   ''', layout = 'constrained')

print(axs.items())

for name, ax in axs.items():
    ax.text(.5, .5, s = f'{name}', va = 'center', ha = 'center')

plt.show()

print(f"Is Interactive? {plt.isinteractive()}")
```
