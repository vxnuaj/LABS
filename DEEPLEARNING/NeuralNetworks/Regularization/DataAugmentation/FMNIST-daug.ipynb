## Augmenting Fashion MNIST Sample

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy as spy
```

### Importing Dataset

```python
data = pd.read_csv('../data/fashion-mnist_train.csv')
data = np.array(data)

X_train = data[:, 1:785].T / 255
Y_train = data[:, 0].reshape(1, -1)
```

### Regular Fashion-MNIST Sample

```python
X_sample = X_train[:, 1].reshape(28, 28)
plt.imshow(X_sample, cmap = 'gray')
```

### Augmented Fashion-MNIST Sample

The image will be augmented as

- Flipping the image
- Zooming out, whislt retaining dimenstionality (28,28)
- Rotating 45° to the right
- Increasing the brightness by 70%

The original np array will be modified, not the plot itself

#### Flipping Axes

```python

X_aug = X_train[:, 1].reshape(28, 28)

# FLIPPING AXES
X_aug = np.flip(X_aug, axis=1)

plt.imshow(X_aug, cmap= 'gray')
```

#### Zooming Out

```python
# ZOOMING OUT
zoom_factor = 0.70
X_aug = spy.ndimage.zoom(X_aug, zoom_factor, prefilter=False)
X_aug = np.pad(X_aug, pad_width = 4 )

plt.imshow(X_aug, cmap= 'gray')
```

#### Rotating Image

```python
# ROTATING IMAGE TO THE RIGHT 45°
X_aug = spy.ndimage.rotate(X_aug, angle = 315, reshape = False)
plt.imshow(X_aug, cmap = 'gray')
```

#### Increasing Brightness

```python
# Increasing brightness
X_aug[X_aug > .5] += .70

print(np.max(X_aug))

plt.imshow(X_aug, cmap = 'gray')
print(X_aug.shape)
```
